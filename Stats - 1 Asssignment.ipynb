```python
"""Problem Statement 5:
A company manufactures LED bulbs with a faulty rate of 30%. If I randomly select 6
chosen LEDs, what is the probability of having 2 faulty LEDs in my sample?
Calculate the average value of this process. Also evaluate the standard deviation
associated with it."""
import math
fault_rate = 0.3
nonfault_rate = 1-fault_rate
n= 6 # number of LEDs selected
fleds = 2 # number of faulty LEDs

# Since our distribution is a binomial distribution for this kind of problem
# we have our solution as nCr*fault_rate**fleds * nonfault_rate**number of non faulty LEDs
print("Final value is ", math.comb(n,fleds)*(fault_rate)**(fleds) * (nonfault_rate)**(n-fleds)) 
# Here our Vaiance = n*p*q and we are taking square of variance calculated
print("Standard deviation for this process is ",(n*fault_rate*nonfault_rate)**0.5) 
```

```python
"""Problem Statement 7:
Customers arrive at a rate of 72 per hour to my shop. What is the probability of 𝑘
customers arriving in 4 minutes? a) 5 customers, b) not more than 3 customers, c)
more than 3 customers. Give a pictorial representation of the same to validate your
answer"""

# For this problem statement, we understood that people are coming at a steady rate on an hourly basis and 
# we need to expand this understanding to minute-wise arrival calculation and so we will proceed forward
# In such cases Poisson's distribution is used
# Since an hour contains 60 mins then per min arrival = 1.2 = 72/60 which is our mean/ mu here just for 1 min
# Then for taking account 4 mins, we will have = 4.8 as mu

from scipy.stats import poisson
import matplotlib.pyplot as plt
import seaborn as sns

probab1 = poisson.pmf(5,4.8)
print("For 5 customers in every 4 mins, the probability is ",probab1)

probab2 = sum(poisson.pmf(k=[0,1,2],mu = 4.8))
print("For not more than 3 customers in every 4 mins , the probability is ", probab2)

print(f'The Probability of more than 3 customers arriving in 4 minutes is : {1-poisson.cdf(k=3,mu=4.8)}')
```

```python
"""Problem Statement 8:
I work as a data analyst in Aeon Learning Pvt. Ltd. After analyzing data, I make
reports, where I have the efficiency of entering 77 words per minute with 6 errors per
hour. What is the probability that I will commit 2 errors in a 455-word financial report?
What happens when the no. of words increases/decreases (in case of 1000 words,
255 words)?
How is the 𝜆 affected?
How does it influence the PMF?
Give a pictorial representation of the same to validate your answer."""

from scipy.stats import poisson

"""Since we know that we have 0.1 errors per min and in 1 min we can type 77 words , therefore we can have 0.1/77 errors
per word"""

unit_mu = 0.1/77
def mu_return(mu_s):
    return unit_mu*mu_s
    
print("Probability of finding 2 error in 455 words financial report", poisson.pmf(2,mu=mu_return(455)))
print("Probability of finding 2 erros in 1000 words financial report", poisson.pmf(2,mu=mu_return(1000)))
print("Probability of finding 2 errors in 255 words financial report", poisson.pmf(2,mu_return(255)))
```

```python
"""Problem Statement 11:
The current flow in a copper wire follow a normal distribution with a mean of 10 𝑚A
and a variance of 4 (𝑚𝐴)2.
What is the probability that a current measurement will exceed 13 𝑚𝐴? What is the
probability that a current measurement is between 9 and 11mA? Determine the
current measurement which has a probability of 0.98.
"""


mean = 10
std = np.sqrt(4) # square of variance is standard deviation

def I(z, b=-np.inf):
    z = (z-mean)/std
    return integrate.quad(norm.pdf,b,z)[0]
print(f"Probability that current > 13mA is: {1-I(13)}")
print(f"Probability that current is between 9 mA and 11 mA is : {1-I(11,b=9)}")
```
